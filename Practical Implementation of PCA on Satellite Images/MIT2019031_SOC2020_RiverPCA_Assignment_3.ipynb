# **Practical Implementation of PCA on Satellite Images**

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
input_mat = np.ndarray(shape=(4, 512*512), dtype=np.integer)

for i in range(4):
    img = Image.open(str(i+1)+'.gif', 'r')
    input_mat[i,:] = np.array(img).flatten()
```

```python
# calculating mean
input_mat = input_mat - np.mean(input_mat)
```

```python
# calculating covariance
cov = np.cov(input_mat)
cov
```

```python
#calculating vectors
_, vec = np.linalg.eig(cov)
```

```python
# calculating principal components
PC = [[],[],[],[]]
for j in range(input_mat.shape[1]):
    for k in range(4):
        PC[k].append(np.dot(vec[:,k], [input_mat[0][j],input_mat[1][j],input_mat[2][j],input_mat[3][j]]))
```

```python
# picking top 2 principal components as output
out_img1 = np.array(PC[0]).reshape((512,512)).astype(np.uint8)
out_img2 = np.array(PC[1]).reshape((512,512)).astype(np.uint8)
```
